```python
import pandas as pd
import joblib
```

```python
model = joblib.load("model.pkl")
```

```python
# input
input_datetime = "2024-04-01 5:00 PM"

dt = pd.to_datetime(input_datetime)
```

```python
X_new = pd.DataFrame([{
    'hour_of_day': dt.hour,
    'day_of_week': dt.dayofweek,
    'day': dt.day,
    'month': dt.month,
    'is_weekend': 1 if dt.dayofweek in [5, 6] else 0
}])
```

```python
# Predict
prediction = model.predict(X_new)[0]

print("Predicted Ride Demand:", round(prediction, 2))
```
